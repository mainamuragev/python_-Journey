```python
import pandas as pd

data = [
    {"name": "Equinix Nairobi", "country": "Kenya", "provider": "Equinix", "power_MW": 5, "status": "Operational"},
    {"name": "Amazon Cape Town", "country": "South Africa", "provider": "AWS", "power_MW": 15, "status": "Operational"},
    {"name": "Google Frankfurt", "country": "Germany", "provider": "Google Cloud", "power_MW": 20, "status": "Under Construction"},
]

df = pd.DataFrame(data)
print(df)
```

```python
summary = df.groupby("provider")["power_MW"].sum().reset_index()
print(summary)
```

```python
import matplotlib.pyplot as plt

plt.bar(summary["provider"], summary["power_MW"], color="skyblue")
plt.title("Total Power Capacity by Provider")
plt.xlabel("Provider")
plt.ylabel("Power (MW)")
plt.show()
```

```python
data = [
    {"name": "Equinix Nairobi", "country": "Kenya", "provider": "Equinix", "power_MW": 5, "status": "Operational", "lat": -1.2921, "lon": 36.8219},
    {"name": "Amazon Cape Town", "country": "South Africa", "provider": "AWS", "power_MW": 15, "status": "Operational", "lat": -33.9249, "lon": 18.4241},
    {"name": "Google Frankfurt", "country": "Germany", "provider": "Google Cloud", "power_MW": 20, "status": "Under Construction", "lat": 50.1109, "lon": 8.6821},
]

import pandas as pd
df = pd.DataFrame(data)
```

```python
import folium

# Create base map centered on Africa
m = folium.Map(location=[0, 20], zoom_start=2)

# Add data center markers
for _, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"{row['name']} ({row['provider']}) - {row['power_MW']} MW",
        tooltip=row['country'],
        icon=folium.Icon(color="green" if row['status'] == "Operational" else "orange")
    ).add_to(m)

m  # Displays the map in Jupyter
```
